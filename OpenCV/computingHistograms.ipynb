```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Fotos/cats.jpg')
```

Histogramas : ajuda a visualizar e analisar a distribuição dos pixels em uma imagem

É possível computar histogramas para imagens RGB ou imagens na escala cinza

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

cv.imshow('Gray', gray)
cv.waitKey()
```

### Histograma de escala cinza
* Ao plotar um histograma, é possível passar uma máscara, afim de mostrar o histograma apenas daquela região específica

```python
# calcHist recebe como parâmetro uma lista de imagens para plotar o histograma
# Tudo é passado como lista
# Onde esta None, poderia ser utilizado uma máscara, assim o histograma seria apenas daquela porção
gray_hist = cv.calcHist([gray], [0], None, [256], [0,256])

plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('Number of pixels')
plt.plot(gray_hist)
plt.xlim([0, 256])
plt.show()
```

* Os bins, parâmetro passado no calculo do histograma, representa os intervalos de valores para intensidade dos pixels

Usar uma máscara ao plotar um histograma pode ser MUITO útil, principalmente quando se deseja analisar partes específicas da imagem.

### Histograma de cores
* Plota as cores em forma de histograma para uma dada imagem

```python
colors = ('b', 'g', 'r')

plt.figure()
plt.title("Color Histogram")
plt.xlabel("Bins")
plt.ylabel("Number of pixels")

for i, col in enumerate(colors):
  hist = cv.calcHist([img], [i], None, [256], [0,256])
  plt.plot(hist, color=col)
  plt.xlim([0, 256])

plt.show()
```
